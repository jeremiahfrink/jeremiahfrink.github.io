```
import matplotlib.pyplot as plt 
import csv

#read data
with open('protein_landuse.csv',newline='') as csvfile:
    data=csv.reader(csvfile,delimiter=',')
    headers = next(data,None)
    xticks=[]
    values_nested=[]
    for row in data:
        xtick=row[0]
        xticks.append(xtick)
        values=[row[1],row[2],row[3],row[4],row[5]]
        values_f = [float(x) for x in values]
        values_nested.append(values_f)

#plot boxplot
plt.boxplot(values_nested,labels=xticks)

#label and rotate xticks for readability
plt.xticks(rotation=90)

#y-axis label 
plt.ylabel('square meters.year') 

#plot title 
plt.title('Land Use per 100g Protein')

#save and display 
plt.savefig('landuse.png', bbox_inches='tight')
plt.show()
```
