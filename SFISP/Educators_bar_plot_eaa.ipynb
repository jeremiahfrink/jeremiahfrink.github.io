```
import matplotlib.pyplot as plt
import csv

#choose the essential amino acid to plot (1-9)
eaa=1

#read data 
with open('essentialaa_food.csv',newline='') as csvfile:
    data=csv.reader(csvfile,delimiter=',')
    headers = next(data,None)
    xticks=[]
    amounts=[]
    for row in data:
        xtick=row[0]
        xticks.append(xtick)
        amount=float(row[eaa])/float(row[10])
        amounts.append(amount)

#confirm the amino acid being read
print(f'Currently reading data for {headers[eaa]}')
```

```
fig = plt.figure()

#bar plot 
plt.bar(xticks,amounts)

#label and rotate xticks for readability
plt.xticks(rotation=70)

#y-axis labels 
plt.ylabel('mg/g protein') 

#plot title 
plt.title(headers[eaa]) 

#display 
plt.show()
```
